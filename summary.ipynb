```python
import matplotlib.pylab as pl
%matplotlib inline
```

```python
methods = ['bmf', 'cbmf'] + ['tbmf-add-%d' % i for i in range(1, 6)]
```

```python
for method in methods:
    x = list(map(float, map(str.strip, open('%s/min' % method).readlines())))
    pl.plot(range(1, len(x) + 1), x, label=method)
pl.plot(range(1, len(x) + 1), [0.803] * len(x), label='full')
pl.grid()
pl.xlabel('t')
pl.ylabel('rmse')
pl.legend(framealpha=0)
```
